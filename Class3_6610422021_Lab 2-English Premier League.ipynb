```python
import pandas as pd

# Load the dataset
url = "https://raw.githubusercontent.com/NattachaiJairak/Football/main/FB.csv"
df = pd.read_csv(url)

# Check for missing data and print the columns with their respective missing counts
missing_data = df.isnull().sum()
print("Missing Data:")
print(missing_data[missing_data > 0])
```

# Which team wins the most as a home team?

```python
# Filter the dataset for home wins
home_wins = df[df['FTR'] == 'H']

# Count the occurrences of each home team in the filtered dataset
most_home_wins_teams = home_wins['HomeTeam'].value_counts()

# Find teams with the maximum number of home wins
max_home_wins = most_home_wins_teams[most_home_wins_teams == most_home_wins_teams.max()]

print("Teams with the most home wins:")
print(max_home_wins)
```

# Which football referee? Who acts as a referee and the home team wins the most and how many times does it win?

```python
# Count the total number of matches officiated by each referee
referee_total_matches = df['Referee'].value_counts()

# Filter the dataset for home wins
home_wins = df[df['FTR'] == 'H']

# Count the number of home wins for each referee
referee_home_wins = home_wins['Referee'].value_counts()

# Find the referee with the most home wins
referee_most_home_wins = referee_home_wins.idxmax()
num_home_wins = referee_home_wins.max()

# Calculate the percentage of games decided for that referee
total_matches_for_referee = referee_total_matches[referee_most_home_wins]
percentage_home_wins = (num_home_wins / total_matches_for_referee) * 100

print(f"The referee who officiates matches with the most home wins is {referee_most_home_wins}.")
print(f"The home team wins {num_home_wins} times.")
print(f"The percentage of games decided is {percentage_home_wins:.2f}%.")
```

# Which Team has most goals?

```python
# Calculate total goals scored at home and away for each team
home_goals = df.groupby('HomeTeam')['FTHG'].sum()
away_goals = df.groupby('AwayTeam')['FTAG'].sum()

# Combine home and away goals to get total goals
total_goals = home_goals.add(away_goals, fill_value=0)

# Identify the team with the most goals
team_most_goals = total_goals.idxmax()
most_goals_scored = total_goals.max()

print("Team that scores the most goals:")
print(f"{team_most_goals} with a total of {most_goals_scored} goals.")
```

# Which team has most corners?

```python
# Calculate total corners for each team at home and away
home_corners = df.groupby('HomeTeam')['HC'].sum()
away_corners = df.groupby('AwayTeam')['AC'].sum()

# Combine home and away corners to get total corners
total_corners = home_corners.add(away_corners, fill_value=0)

# Identify the team with the most corners
team_most_corners = total_corners.idxmax()
most_corners_count = total_corners.max()

print("Team with the most corners:")
print(f"{team_most_corners} with a total of {most_corners_count} corners.")
```

# Which team has the highest away win rate?

```python
# Filter the dataset for away wins
away_wins = df[df['FTR'] == 'A']

# Calculate the total number of away matches played by each team
total_away_matches = df['AwayTeam'].value_counts()

# Calculate the total number of away wins for each team
away_wins_count = away_wins['AwayTeam'].value_counts()

# Calculate the away win rate for each team
away_win_rate = (away_wins_count / total_away_matches).sort_values(ascending=False)

# Identify the team with the highest away win rate
team_highest_away_win_rate = away_win_rate.idxmax()
highest_away_win_rate = away_win_rate.max()

print("Team with the highest away win rate:")
print(f"{team_highest_away_win_rate} with an away win rate of {highest_away_win_rate:.2%}")
```
